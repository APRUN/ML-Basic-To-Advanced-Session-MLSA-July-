## For Term Insurance Dataset

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv('TermInsuranceData.csv')
df
```

```python
sns.lmplot(x='Age', y='Premium',data=df)
```

```python
reg=linear_model.LinearRegression()
#first element should be 2d
reg.fit(df[['Age']],df['Premium'])
```

```python
reg.predict([[21]])
```

```python
m=reg.coef_
m
```

```python
c=reg.intercept_
c
```

# With Multiple Variables:

```python
import pandas as pd
from sklearn import linear_model

df=pd.read_csv('TermInsuranceData2.csv')
df
```

```python
print(df.Height)
print(df.Height.mean())
df.Height=df.Height.fillna(df.Height.mean())
```

```python
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Age','Height','Weight']],df['Premium'])
```

```python
reg.predict([[25,162.56,70]])
```
